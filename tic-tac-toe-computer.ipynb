# __Tic Tac Toe__

### __Reglas del Juego "Tic-Tac-Toe":__

- El juego se juega en un tablero de 3x3.
- Dos jugadores, uno representa "X" y el otro "O", alternan turnos.
- El objetivo es obtener tres símbolos del mismo tipo en línea horizontal, vertical o diagonal.
- El juego termina cuando un jugador gana o el tablero se llena (empate).
- El jugador que logre colocar tres de sus símbolos en línea es el ganador.

```python
def welcome_1():
    print("Welcome to Tic Tac Toe, a game from Gaming S.L.! 🎮 Get ready to challenge your mind.\n")
    print("Your mission:\n")
    print("Outsmart your opponent by being the first to align three of your marks in a row, column, or diagonal.\n")
    print("How to play:\n")
    print("1. Players take turns placing their mark (❌ or ⭕) in an empty spot.")
    print("2. Once a spot is marked, it cannot be used again.")
    print("3. Keep going until a player wins or the board is full. If no one wins… it’s a draw!\n")
    print("Obs! If you do not type a real number you will be kicked out of the game!")
    print("Are you friendless? Choose to play against fate! (i.e. your friend the computer)\n")
    return 

def welcome_2():
    print("Let's start the game!\n")
    print("Here’s your board. Make your move wisely:\n")
    return



def board_view():
        print(f" ―――――――――――――\n",
              "|", board[0][0],"|", board[0][1],"|", board[0][2],"|\n", 
              "―――――――――――――\n",
              "|", board[1][0],"|", board[1][1],"|", board[1][2],"|\n", 
              "―――――――――――――\n",
              "|", board[2][0],"|", board[2][1],"|", board[2][2],"|\n", 
              "―――――――――――――")
        return
```

```python
def x_place_marker():

    updated = False

    while updated == False:

            marker_x = int(input("Player 1. Please, place your 'X' by choosing an available spot from 1 to 9. Be careful! If you do not type a real number you will be kicked out of the game!"))
            
            if (marker_x >= 1) and (marker_x <= 9) and (marker_x != 99):

                for row in range(3): #Si no nos metemos en las listas de la lista principal solo buscará el número en la primera sublista
                    for column in range(3): #Con range 3 accedemos a cada elemento de cada fila y cada columna
                        if board[row][column] == marker_x:
                            updated = True
                            board[row][column] = "❌"
                            board_view()
                            break
            
                if not updated:
                    print("Player 1: The selected spot is not available. Please, choose a diferent one")     

            else: 
                print("Player 1: Please, be careful! Choose a number inside the range 1 to 9")   
    
    return    


def o_place_marker():

    updated = False

    while updated == False:

            marker_o = int(input("Player 2. Please, place your 'O' by choosing an available spot from 1 to 9. Be careful! If you do not type a real number you will be kicked out of the game!"))

            if (marker_o >= 1) and (marker_o <= 9):

                for row in range(3):
                    for column in range(3):
                        if board[row][column] == marker_o:
                            updated = True
                            board[row][column] = "⭕"
                            board_view()
                            break

                if not updated:
                    print("Player 2: The selected spot is not available. Please, choose a diferent one")

            else: 
                print("Player 2: Please, be careful! Choose a number inside the range 1 to 9")

    return


def o_place_marker_pc():

    updated = False

    while updated == False:

            choices = [1,2,3,4,5,6,7,8,9]

            import random
            marker_o_pc = random.choice(choices)

            for row in range(3):
                for column in range(3):
                    if board[row][column] == marker_o_pc:
                        updated = True
                        board[row][column] = "⭕"
                        board_view()
                        break

    return
```

```python
def control():

    game_over = False

    while game_over == False:

        #horizontal
        if board[0][0] == board[0][1] == board[0][2]:
            if board[0][0] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True

        elif board[1][0] == board[1][1] == board[1][2]:
            if board[1][0] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True

        elif board[2][0] == board[2][1] == board[2][2]:
            if board[2][0] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True

        # vertical
        elif board[0][0] == board[1][0] == board[2][0]:
            if board[0][0] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True

        elif board[0][1] == board[1][1] == board[2][1]:
            if board[0][1] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')        
            game_over = True

        elif board[0][2] == board[1][2] == board[2][2]:
            if board[0][2] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True

        # diagonal
        elif board[0][0] == board[1][1] == board[2][2]:
            if board[0][0] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True

        elif board[0][2] == board[1][1] == board[2][0]:
            if board[0][0] == '❌':
                player = 'Player 1'
            else:
                player = 'Player 2'
            print(f'Game done, {player} has won!')
            game_over = True
            
        # tablero lleno sin ganador 
        else:
            all_spots = board[0] + board[1] + board[2]
            all_strings = True

            for spot in all_spots:
                if spot in  range(1,10):
                    all_strings = False
                    game_over = False
                    break
            
            if all_strings:
                game_over = True
                print('All spots are filled and no player has won. It’s a draw!')
            else:
                print('Keep playing')
        
        return game_over
```

```python
def restart_game():
    # Preguntar si quieren jugar de nuevo
    answer = input("Do you want to play again? (yes/no):")
    if answer.lower() == "yes":
        print("\nStarting a new game...\n")
        start_game()  # Llamamos a la función principal que inicia el juego (definir el nombre)
    else:
        print("\nThanks for playing Tic Tac Toe! Goodbye! 👋")
        
    return board # Simplemente termina la función sin cerrar el programa
    
# si usamos esta función, podemos definir una que sea iniciar_juego() y llamarla en lugar de bienvenida() para que el juego se reinicie automáticamente
```

```python

def start_game():

    welcome_1()  # Mostramos el mensaje de bienvenida
    mode = int(input("Please, type (1) to play against the computer or (2) to play with your real friend"))
    welcome_2()

    keep_playing = True
    game_over = False
    board.clear()
    board.append([1,2,3])
    board.append([4,5,6])
    board.append([7,8,9])
    board_view()  # Mostramos el tablero vacío

    if mode == 1:

           while keep_playing == True:        

            if game_over == False:

                x_place_marker()
                game_over = control()

                if game_over == False:
                    o_place_marker_pc()
                    game_over = control()

                if game_over:  
                    keep_playing = False
                    board.clear()
                    board.append([1,2,3])
                    board.append([4,5,6])
                    board.append([7,8,9])
                    print("Game Over")
                    restart_game()
                    break
                
    elif mode == 2:

        while keep_playing == True:        

            if game_over == False:

                x_place_marker()
                game_over = control()

                if game_over == False:
                    o_place_marker()
                    game_over = control()

                if game_over:  
                    keep_playing = False
                    board.clear()
                    board.append([1,2,3])
                    board.append([4,5,6])
                    board.append([7,8,9])
                    print("Game Over")
                    restart_game()
                    break

    else: 
        print("Please, choose a valid input. Type (1) to play against the computer or (2) to play with your real friend")
        
    return
```
